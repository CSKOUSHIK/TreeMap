```python
d=sns.load_dataset('diamonds')   #Loading Diamonds Dataset
d.head()
```

```python
d.dropna(inplace=True)
```

```python
k=d['color'].value_counts()
```

```python
squarify.plot(sizes=k.values,label=k.index); #A Basic Tree Map
plt.axis('off')
```

```python
u=d['cut'].value_counts()
u
```

```python
i=u/u.values.sum()
```

```python
i
```

```python
i.index[0]
```

```python
p=0
j=[]
for w in i:
    j.append(i.index[p]+'('+str(round(w,4))+')')
    p=p+1
```

```python
j
```

```python
plt.figure(figsize=(14,8))
squarify.plot(sizes=u.values,label=j,pad=2,color=['red','green','goldenrod','deepskyblue','violet'],text_kwargs={'fontsize':'x-large','fontweight':'bold','color':'white'});
plt.axis('off')
plt.title('Tree Diagram')
# plt.label(fontsize='large')
```

```python
p=0
j=[]
for w in u:
    j.append(i.index[p]+'('+str(w)+')')
    p=p+1
```

```python
plt.figure(figsize=(14,8))
squarify.plot(sizes=u.values,label=j,pad=2,color=['red','green','goldenrod','deepskyblue','violet'],text_kwargs={'fontsize':'x-large','fontweight':'bold','color':'white'});
plt.axis('off')
plt.title('Tree Diagram')
```

```python
plastic=pd.read_csv('plastic.csv')
plastic
```

```python
plastic.dropna(inplace=True)
```

```python
plastic.shape
```

```python
plastic.sort_values(by='Total mismanaged plastic waste in 2025',ascending=False,inplace=True)
```

```python
plastic
```

```python
plastic[['Entity','Total mismanaged plastic waste in 2025']]
```

```python
plastic['% of Waste in 2025']=(plastic['Total mismanaged plastic waste in 2025']/plastic['Total mismanaged plastic waste in 2025'].sum())*100
```

```python
plastic.reset_index(drop=1)
# plastic.drop('index',axis=0,inplace=True)
```

```python
p=0
j=[]
for w in plastic['% of Waste in 2025'].head(15):
    j.append(plastic['Entity'][plastic['% of Waste in 2025'].head(15).index[p]]+'('+str(round(w,2))+'%'+')')
    p=p+1
```

```python
plastic['% of Waste in 2025'].head(15)
```

```python
j
```

```python
plt.figure(figsize=(32,18))
squarify.plot(sizes=plastic['Total mismanaged plastic waste in 2025'].head(15),label=j,color=['red','deepskyblue','orange','HotPink','purple','lawngreen','chocolate','mintcream','Silver']
             ,text_kwargs={'fontsize':35});
plt.tight_layout()
plt.axis('off')
plt.title('TREEMAP - Estimation of total percentage of mismanaged plastic waste in 2025 by each country',fontsize=45,color='teal');
```

```python
happy=pd.read_csv('Happy.csv')
happy.head(11)
```

```python
twenty_eighten=happy[happy['Year']==2018]
twenty_eighten
```

```python
twenty_eighten.sort_values('Life satisfaction in Cantril Ladder (World Happiness Report 2018)',ascending=False,inplace=True)
twenty_eighten
```

```python
twenty_eighten['Entity']
```

```python
twenty_eighten.reset_index(drop=1)
```

```python
twenty_eighten['Entity'].head(10)
```

```python
p=0
y=[]
for j in range(1,11):
    y.append(twenty_eighten['Entity'].head(10).values[p] +'('+str(j)+')')
    p=p+1
```

```python
plt.figure(figsize=(14,8))
squarify.plot(sizes=twenty_eighten['Life satisfaction in Cantril Ladder (World Happiness Report 2018)'].head(10),
              color=['red','HotPink','deepskyblue','orange','purple','green','chocolate','mintcream','Silver','gold'],label=y,
             text_kwargs={'fontsize':'x-large'})
plt.axis('off')
plt.title('Top 10 Countries in World Happiness Report 2018',fontsize='xx-large',color='coral')
plt.savefig('WHC')
```
